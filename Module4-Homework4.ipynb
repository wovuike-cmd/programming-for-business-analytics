# Question 1

```python
import pandas as pd

def check_all_odd_pandas(L):
    #converts the list into a panda Series
    s = pd.Series(L)

    #check all value on list for odd number using all() function
    result = (s % 2 != 0).all()
    print(result)

#gets three seperate inputs
user_input = input("Please input 3 separated numbers: ")

#turns each input into a integer
L1 =  list(map(int, user_input.split()))
check_all_odd_pandas(L1)
```

# Question 2

```python
def smallest_even_odd(num_list):
    even_list = []
    odd_list = []

    for num in num_list: 
        #check if num is even or odd
        if num % 2 == 0:
            #if num is even the value in num is inserted into even variable
            even_list.append(num)
        else:
            # else num is inserted into odd variable
            odd_list.append(num)

     #using the sort() function to sort the list from smallest to largest
    even_list.sort()
    odd_list.sort()

    #checking if list is empty 
    if len(even_list) > 0:
        print("Smallest even number:", even_list[0])
    else:
        print("No even numbers in the List")

    if len(odd_list) > 0:
        print("Smallest odd number:", odd_list[0])
    else:
        print("No odd numbers in the list")

#getting input for list
user_input = input("Please input 3 separated numbers: ")
given_list = list(map(int, user_input.split()))
smallest_even_odd(given_list)
```

```python
given_list = [11,23,58,31,56,22,43,12,65,19]
n = 3

evens = [] 

for num in given_list:
    if num % 2 == 0:
        evens.append(num)

# removing the smallet observation
for i in range(n -  1):
    #finding the current miniumum even number
    current_min = evens[0]
    for val in evens:
        if val < current_min:
            current_min = val

    evens.remove(current_min)

#Find the minimum observation of the updated list 
final_nth_min = evens[0]
for val in evens:
    if val < final_nth_min:
        final_nth_min = val

print(f"The {n} smallest even value is: {final_nth_min}")
```
