# Question 1

```python
#Random number between 0 to 1
u = np.random.rand()
v = np.random.rand()
#Finding v
sine_v = np.sin(2 * np.pi * v)
#Finding u
sqrt_u = np.sqrt(-2 *  np.log(u))
#Finding Z
z_equation = sine_v * sqrt_u
print("Sine v:", sine_v)
print("Sqaure Root u:", sqrt_u)
print("The Value of Z:",z_equation)
```

```python
data_size = df['Descriptor'].size
data_min = df["Magnitude-min"]
data_max = df["Magnitude-max"]
# print(float(data_min[0]))
# print(float(data_max[0]))
user_input = float(input("Please enter first input: "))
i = 0
while i < data_size:
    if float(data_min[i]) <= user_input <= float(data_max[i]):
        print("Decriptor: ", df['Descriptor'][i])
        break
    elif user_input >= float(df["Magnitude-max"][8]):
        print("Decriptor: ", df['Descriptor'][8])
        break
    else :
        i += 1
```
