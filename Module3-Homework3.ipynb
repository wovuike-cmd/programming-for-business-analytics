# Question 1 

```python
first_num = 150
last_num = 1000

count_odd = 0 
sum_odd = 0 
multiplication_odd = 1 
sum_square_odd = 0

while first_num <= last_num:
    if first_num % 2 != 0:
        count_odd += 1
        sum_odd += first_num
        multiplication_odd *= first_num
        sum_square_odd += (first_num **2)
    first_num += 1

print("1a Number of odd numbers:", count_odd)
print("1b Sum of odd numbers:", sum_odd)
print("1c Multiplication of odd numbers:", multiplication_odd)
print("1d Sum of sqaures of odd numbers:", sum_square_odd)
```

# Question 2

```python
first_num = 150
last_num = 1000

count_odd = 0 
sum_odd = 0 
multiplication_odd = 1 
sum_square_odd = 0

for num in range(first_num, last_num):
    if num% 2 !=0:
        count_odd += 1
        sum_odd += num
        multiplication_odd *= num
        sum_square_odd += (num ** 2)
print("1a Number of odd numbers:", count_odd)
print("1b Sum of odd numbers:", sum_odd)
print("1c Multiplication of odd numbers:", multiplication_odd)
print("1d Sum of sqaures of odd numbers:", sum_square_odd)
```

# Question 3

```python
m_input = int(input("input number for M"))

s = 0.0
k = 1

while k <= m_input:
    s = s + (1/k)
    k = k + 1

print(f"The sum for M={m_input} is: {s}")
```

# Question 4

```python
total_sum = 0
count = 0

while True:
    user_input = input("Enter a number")

    if user_input == "done":
        break
    try:
        num = float(user_input)

        total_sum += num 
        count += 1 

    except ValueError:
     print("Enter a number: bad data Invalid input")

if count > 0:
    average = total_sum / count
    print(total_sum, count, average)

else:
    print("No numbers were entered.")
```

# Question 5

```python
maximum = None
minimum = None

while True:
    user_input = input("Enter a number: ")
    if user_input == "done":
        break
    try:
        num = float(user_input)
        if maximum is None:
            maximum = num
            minimum = num
        else:
            if num > maximum:
                maximum = num

            if num < minimum: 
                minimum = num
                
    except ValueError:
        print("Invalid input")
if maximum is not None:
    print("Maximum:", maximum)
    print("Minimum:", minimum)
else:
    print("No number were entered.")
```
